### Cálculo de segundos em 42 minutos e 42 segundos

Para calcular o total de segundos, multiplicamos os minutos por 60 (para converter em segundos) e adicionamos os segundos restantes.

```python
# Definindo os valores
minutos = 42
segundos_adicionais = 42

# Calculando o total de segundos
total_segundos = (minutos * 60) + segundos_adicionais

# Exibindo o resultado
print(f"Em {minutos} minutos e {segundos_adicionais} segundos, existem {total_segundos} segundos.")
```

### Quantas milhas existem em 10 quilômetros

Para calcular o total de milhas, dividimos os quilômetros por 1.61 (para converter em milhas).

```python
# Definindo valores
quilometros = 10

# Calculando o total de milhas
total_milhas = quilometros / 1.61

# Exibindo o resultado
print(f"{quilometros} quilômetros equivalem a {total_milhas} milhas.")
```

### Se você correr 10 quilômetros em 42 minutos e 42 segundos, qual será o seu passo médio(tempo por milhasem minutos e segundos)?

Para calcular o total de milhas, dividimos os quilômetros por 1.61 (para converter em milhas). Precisa dividir os segundos por 60 para trasformar os 42 segundos em minutos e somar o total de minutos. Para o total de segundos multiplicar 42 minutos por 60 e somar aos 42 segundos adicionais.

```python
# Definindo valores
minutos_corrida = 42
segundos_adicionais_corrida = 42
quilometros = 10

# Calculando o tempo total da corrida em minutos
total_tempo_em_minutos = minutos_corrida + (segundos_adicionais_corrida / 60)

# Calculando o tempo total da corrida em segundos
total_tempo_em_segundos = (minutos_corrida * 60) + segundos_adicionais_corrida

# Calculando o total de milhas
total_milhas = quilometros / 1.61

# Calculando o passo médio em segundos por milha
passo_segundos_por_milha_float = total_tempo_em_segundos / total_milhas

# Convertendo o passo médio (segundos por milha) para minutos e segundos
passo_minutos_inteiro = int(passo_segundos_por_milha_float // 60)
passo_segundos_restante = passo_segundos_por_milha_float % 60

# Calculando a velocidade média em milhas por hora
total_tempo_em_horas = total_tempo_em_minutos / 60
velocidade_media_mph = total_milhas / total_tempo_em_horas

# Exibindo resultados
print(f"Em {minutos_corrida} minutos e {segundos_adicionais_corrida} segundos, você correu {quilometros} quilômetros.")
print(f"Isso equivale a {total_milhas:.2f} milhas.")
print(f"Seu passo médio foi de {passo_minutos_inteiro} minutos e {passo_segundos_restante:.2f} segundos por milha.")
print(f"Sua velocidade média foi de {velocidade_media_mph:.2f} milhas por hora.")
```
