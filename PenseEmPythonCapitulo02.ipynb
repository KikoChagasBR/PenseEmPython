#### O volume de uma esfera com raio r é 4/3*pi*r**3. Qual seria o volume de uma esfera com raio 5?

```python
# Importar módulo math
import math

# Definindo valores
radius = 5

# Calculando o volume
volume = (4/3) * math.pi * (radius**3)

# Exibindo o resultado
print(f"O volume da esfera com raio {radius} é: {volume}")
```

#### Existe uma regra na trigonometria que diz que, para qualquer valor de x, (cos x)**2 + (sin x)**2 = 1. Vamos verificar se isso é verdade para um valor específico de x, como 42.

```python
# Importar módulo math
import math

# Definindo valores
x = 42

# Verificando a regra trigonométrica
result = (math.cos(x)**2) + (math.sin(x)**2)

# Exibindo o resultado
print(f"O resultado da regra trigonométrica para x = {x} é: {result}")
```

#### Além de pi, outra variável definida no módulo math é e, que representa a base do logaritmo natural, escrita na notação matemática como e. Vamos calcular de 3 formas distintas:
1. Utilize math.e e operador de exponenciação (**)
2. Utilize math.pow para elevar math.e à potência 2.
3. Utilize math.exp que recebe como argumento um valor, x, e calcule e**x.

```python
# Importar módulo math
import math

# Definir valores
potencia = 2

# Usar math.e com exponenciação (**)
resultado1 = math.e ** potencia
print(f"Usando math.e com exponenciação (**): {resultado1}")

# Usar math.pow para elevar math.e à potência 2
resultado2 = math.pow(math.e, potencia)
print(f"Usando math.pow com potência 2: {resultado2}")

# Usar math.exp que recebe como argumento
resultado3 = math.exp(potencia)
print(f"Usando math.exp com potência 2: {resultado3}")
```
