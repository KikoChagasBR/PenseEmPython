#### Escreva uma função chamada print_rigth, que recebe uma string chamada text como parâmetro e exiba essa string com espaços suficientes à frente para que a última letra esteja alinhada na 40 coluna da tela.

```python
# Definição da função
def print_rigth(text):
    print(text.rjust(40))

# Usando a função com exemplos de texto fornecidos pelo livro
print_rigth("Monty")
print_rigth("Python's")
print_rigth("Flying Circus")
```

#### Escreva uma função chamada triangle que receba uma string e um número inteiro, e desenhe um triângulo, composto de múltiplas cópias da string.

```python
# Definindo a função e aplicando a mesma com uso do loop for
def triangle(text, n):
  for i in range(n + 1):
    print(text * i)

# Usando a função criada usando o texto e números sugeridos no livro
triangle("L", 5)
```

#### Escreva uma função chamada rectangle que receba uma string e dois números inteiros e desenhe um retângulo com a largura e altura fornecidas, composto de cópias da string.

```python
# Definindo a função e aplicando o loop
def rectangle(text, x, y):
  for i in range(y):
    print(text * x)

# Usando a função criada usando o texto e números sugeridos no livro
rectangle("H", 5, 4)
```

#### A música "99 Bottles of Beer" começa com este verso:
  99 bottles of beer on the wall,
  99 bottles of beer.
  Take one down, pass it around,
  98 bottles of beer on the wall.

Em seguida, o segundo verso é semelhante, exceto que começa com 98 garrafas e termina com 97. a música continua ---- por um longo tempo ---- até que não reste nenhuma garrafa de cerveja (bottles of beer).

Escreva uma função chamada bottle_verse que receba um número como parâmetro e exiba o verso que começa com o número de garrafas fornecido.

```python
# Definindo a função
def bottle_verse(n):
  print(f"{n} bottles of beer on the wall,")
  print(f"{n} bottles of beer.")
  print(f"Take one down, pass it around,")
  print(f"{n - 1} bottles of beer on the wall.")

# Crianda a versão com 99 garrafas como sugerido no livro
bottle_verse(99)
```

```python
# Versão que usa o loop for para apresentar a música toda de 1 até 99
def bottle_verse(n):
  for i in range(n, 0, -1):
    print(f"{i} bottles of beer on the wall,")
    print(f"{i} bottles of beer.")
    print(f"Take one down, pass it around,")
    print(f"{i - 1} bottles of beer on the wall.")

bottle_verse(99)
```
