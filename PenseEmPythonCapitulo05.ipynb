#### Este código contém um erro, pergunte a um assistente virtual qual o erro e como corrigir.

```python
def countdown_by_two(n):
  if n == 0:
    print("Blastoff!")
  else:
    print(n)
    countdown_by_two(n-2)
```

```python
countdown_by_two(6)
```

#### O erro esta na segunda linha do código:
"if n == 0:", enquanto n for um número par a chamada countdown_by_two(n-2), não vai gerar problemas, pois sempre chegaremos exatamente no '0' (zero) mas no caso de um número ímpar nunca chegaremos exatamente no '0' (zero), o que vai gerar uma recursão infinita. Para resolver temos os 2 cenários abaixo como sugestão:

```python
# Sugestão 1 - Parar quando n <= 0
def countdown_by_two(n):
    if n <= 0:
        print("Blastoff!")
    else:
        print(n)
        countdown_by_two(n - 2)
```

```python
# Testando a sugestão 1
countdown_by_two(10)
countdown_by_two(9)
```

```python
# Sugestão 2 - Garantir entrada válida (mais inflexível)
def countdown_by_two(n):
    if n % 2 != 0:
        raise ValueError("n must be even")
    if n == 0:
        print("Blastoff!")
    else:
        print(n)
        countdown_by_two(n - 2)
```

```python
# Testando a sugestão 2
countdown_by_two(10)
countdown_by_two(9)
```

#### O módulo time oferece uma função, também chamada time, que retorna o número de segundos desde o Unix/Epoch, que corresponde a 1 de janeiro de 1970, 00:00:00 UTC (Tempo Universal Coordenado):
Utilize a divisão pelo piso e a operação de módulo para calcular o número de dias desde 1 de janeiro de 1970 e o horário atual em horas, minutos e segundos

```python
from time import time

now = time()

SECONDS_PER_MINUTE = 60
SECONDS_PER_HOUR = 60 * 60
SECONDS_PER_DAY = 24 * SECONDS_PER_HOUR

days = int(now // SECONDS_PER_DAY)

remaining_seconds = int(now % SECONDS_PER_DAY)

hours = (remaining_seconds // SECONDS_PER_HOUR) - 3
remaining_seconds %= SECONDS_PER_HOUR

minutes = remaining_seconds // SECONDS_PER_MINUTE
seconds = remaining_seconds % SECONDS_PER_MINUTE

print("Days since Jan 1, 1970:", days)
print("Current time:", hours, ":", minutes, ":", seconds)
```

#### Escreve uma função is_triangle que receba três números inteiros como argumentos, e que exiba "Yes" ou "No", dependendo da possibilidade de formar ou n'ao triângulo de graveto com os comprimentos fornecidos. Dica: utilize uma condicional encadeada.

teste para verificar se é possível formar um triângulo:
Se qualquer um dos tres que comprimentos for maior que a soma dos outros dois, então não é possível formar um triângulo. Caso contrário, é possível. (Se a soma de dois comprimentos for igual ao terceiro, eles formam o que é chamdo de triângulo "degenerado".)

```python

def is_triangle(a, b, c):
  if a > b + c or b > a + c or c > a + b:
    print("No")
  elif a == b + c or b == a + c or c == a + b:
    print("No")
  else:
    print("Yes")
```

```python
is_triangle(4, 5, 6)
```

```python
is_triangle(1, 2, 3)
```

```python
is_triangle(6, 2, 3)
```

```python
is_triangle(1, 1, 12)
```

#### Qual a saída do programa a seguir? Desenhe um diagrama de pilha que mostre o estado do programa quando exibir o resultado.

```python
def recurse(n, s):
    if n == 0:
        print(s)
    else:
        recurse(n-1, n+s)

recurse(3, 0)
```
